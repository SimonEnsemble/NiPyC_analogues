# Plot Adsorption Isotherms

In this notebook:
1. Henry coefficient fits to experimental isotherm data
2. Experimental and Simulated adsorption isotherms for:

    (i)  Ni(PyC)2 
        (a) Xe, Kr, Ar @ T = 298 K
    (ii) Ni(PyC-m-NH2)2 
        (a) Xe, Kr, Ar @ T = 298 K 
        (b) Xe in different topologies: 
            - single-net v.s. double-net 
            - Pn v.s. P2_1/n (cis v.s. trans PyC)
3. Experimental (only) isotherms:

    (i)  Ni(PyC)2
         (a) Xe, Kr, and Ar @ T = 278 K, 288 K, 298 K 
         (b)Nitrogen adsorption and desorption @ T = 77 K
    (ii) Ni(PyC-m-NH2)2
         (a) Xe, Kr, and Ar @ T = 278 K, 288 K, and 298 K 
         (b) Nitrogen adsorption and desorption @ T = 77 K
4. DFT relaxation methods comparison plot (Xe adsorption @ T = 298 K)

```julia
###
#  properties we want to pull from the simulation results dictionary
###
props_we_want = ["pressure (bar)", "fugacity (bar)", "⟨N⟩ (mmol/g)"]

###
#  Gas (adsorbate) molecules:
###
adsorbates = ["Xe", "Kr", "Ar"]
molar_mass  = Dict("Xe" => 131.293, "Kr" => 83.798, "Ar" => 39.948) # [g / mol], used for unit conversions

###
#  Crystal Structures:
#   unless specified, this is the default (Parent MOF) structure we will analyze
###
parent_mofs = ["NiPyC2"]

###
#  Functional Groups:
#   unless specified, these are the default functional groups we will analyze
#   N.B. the functioal group "" is just the Parent MOF (i.e. no functional group)
###
functional_groups = ["", "mNH2"]

###
#  simulation parameters
#   used to locate desired simulation files
###
pmin  = -3  # minimum pessure in log10, units: bar
pmax  = 1.0 # maximum pressure (actual value), units: bar
nstep = 25  # number of intervals in pressure range

sim_params = Dict("adsorbates"  => adsorbates,
                  "temperature" => 298.0, # units: K
                  "molecules"   => Dict(zip(adsorbates, Molecule.(adsorbates))),
                  "pressures"   => 10 .^ range(pmin, stop=log10(pmax), length=nstep),
                  "ljff"        => LJForceField("UFF"), # molecular force field model
                  "n_burn"      => 50000, # number of burn cycles in simulation
                  "n_sample"    => 50000  # number of production cycles 
                  )
```

## Functions for Data Retrieval and Fitting

### Experimental Data

```julia
###
#  Unit conversion dictionary
#  we want to report and plot: ⟨N⟩ (mmol/g)" vs. pressure (bar)
###
# 1000 mbar = 1 bar
pressure_conversion = Dict("P(mbar)" => 1 / 1000)

# (cc STP /g) * (1000 g /kg) * (1 mol/ 22.4 L STP) * (1 L / 1000 cc)
# [(% mass) / 100 g gas / g MOF](1 mol / MW_gas g)
loading_conversion = Dict("loading (ccSTP/g)" => 1 / 22.4,
                          "PercentMass" => Dict(zip(adsorbates, 
                                                [1000.0 / molar_mass[gas] / 100.0 for gas in adsorbates]))
                          )
```

```julia
function unit_conversion!(data::DataFrame, adsorbate::String)
    # make a copy of input DataFrame
    df = deepcopy(data)
    
    # apply unit conversion
    for col_name in names(df)
        # check adsorption units
        if col_name in keys(loading_conversion)
            # special case
            if col_name == "PercentMass"
                df[:, Symbol("loading (mmol/g)")] = df_data[:, col_name] * 
                                                     loading_conversion[col_name][adsorbate]
            end
            df[:, Symbol("loading (mmol/g)")] = df[:, col_name] * loading_conversion[col_name]
        # check pressure units
        elseif col_name in keys(pressure_conversion)
            df[:, Symbol("pressure (bar)")] = df[:, col_name] * pressure_conversion
        end
    end
    # rename to adsorption column for consistency 
    rename!(df, Symbol("loading (mmol/g)") => Symbol("⟨N⟩ (mmol/g)"))
    
    return df
end
```

```julia
"""
    df = experimental_isotherm_data(adsorbate, xtal_name, fgroup)

# Arguments
- `adsorbate::String`: the gas molecule for which the data was taken
- `xtal_name::String`: name of the crystal structure which the data was taken for
- `fgroup::String`="": the functional group, if any, decorating the ligand
- `temperature::Int64=298`: temperature [K] of the isotherm
- `ignore_desorption::Bool=true`: do not include desorption data in dataframe
- `add_origin::Bool=true: add the point (0, 0) to the dataframe
- `data_loc::String=""`: location of data files if not in joinpath(PorousMaterials.PATH_TO_DATA,
                         "experimental_isotherm_data", "xtal_name")

# Returns 
- `df::DataFrame`: a `DataFrame` containing the experimental data  
"""
function experimental_isotherm_data(adsorbate::String, xtal_name::String; 
                                    fgroup::String="", temperature::Int64=298, 
                                    ignore_desorption::Bool=true, 
                                    add_origin::Bool=true, data_loc::String="")
    # define filename with the convention:
    # XtalName-fgroup_T_Temperature.csv
    filename = xtal_name
    if ! isempty(fgroup)
        filename *= "-" * fgroup
    end
    filename *= "_" * adsorbate* "_T_$(temperature).csv"
    # define path to data file
    file_path = joinpath(PorousMaterials.PATH_TO_DATA, "experimental_isotherm_data", xtal_name, filename)
    
    # load data from .csv file into a DataFrame
    df_data = CSV.read(file_path, DataFrame)
    
    # apply unit conversion 
    df_data = unit_conversion!(df_data, adsorbate)
    
    # ignore desortion
    if ignore_desorption
        desorp_onset = argmax(df_data[!, Symbol("pressure (bar)")]) # onset of desorption
        df_data = df_data[1:desorp_onset, [Symbol("pressure (bar)"), Symbol("⟨N⟩ (mmol/g)")]]
    end
    
    # add a point at the origin
    if add_origin
        push!(df_data, zeros(length(names(df_data))))
    end
    
    # sort in order of increasing pressure
    if ignore_desorption
        sort!(df_data, Symbol("pressure (bar)"))
    end
    return df_data
end
```

```julia
# store experimental data in a dictionary for easy access
exptl_data = Dict{Tuple{String, String, String}, DataFrame}()
for mof in parent_mofs
    for fgroup in functional_groups
        for gas in adsorbates
            exptl_data[(mof, fgroup, gas)] = experimental_isotherm_data(gas, mof; fgroup=fgroup)
        end
    end
end
```

### Simulation Data

```julia
"""
    df = function read_sim_data(props, xtal, gas, temperature, ljff, pressures, nburn, nsample)

This function calls [`isotherm_sim_results_to_dataframe`](@ref) under the hood.

# Arguments
- `props::Array{String, 1}`:
- `xtal::Crystal`:
- `gas::Molecule{Cart}`:
- `temperature::Float64`:
- `ljff::LJForceField`:
- `pressures::Array{Float64}`:
- `nburn::Int64`:
- `nsample::Int64`:

# Returns
- `df`: a `DataFrame` containing the specified properties 
"""
function read_sim_data(props::Array{String, 1}, xtal::Crystal, 
                       gas::Molecule{Cart}, temp::Float64, ljff::LJForceField, 
                       pressures::Array{Float64}, nburn::Int64, nsample::Int64)
    press = [[p] for p in pressures]
    df_sim = isotherm_sim_results_to_dataframe(props, xtal, [gas], temp, press, ljff, nburn, nsample)
    push!(df_sim, [[0.0], [0.0], [0.0]]) ### add point at origin ###
    sort!(df_sim, Symbol("pressure (bar)"))
    return df_sim
end

# create a second method that is easier to implement
function read_sim_data(props::Array{String, 1}, params::Dict{String, Any}, gas::String; xtal::String="")
    # check that a crystal was passed 
    if ! isempty(xtal)
        params["xtal"] = Crystal(xtal)
    else
        @assert ! isempty(params["xtal"])
    end
    # check that gas is in adsorbates
    @assert (gas in params["adsorbates"])
    mol = Molecule(gas)
    return read_sim_data(props, params["xtal"], mol, params["temperature"],
                         params["ljff"], params["pressures"], params["n_burn"], params["n_sample"])
end

###
#  account for change in unit cell volume when plotting adsorbates/unit_cell comparison
###
function adsorption_scale_factor(mof::String)
    xtal = Crystal(mof)
    strip_numbers_from_atom_labels!(xtal)
    ## conversion:
    # N_A = 6.0221409e+23 atoms gas / mol gas
    # crystal density [ρ] = kg mof/m³
    # [xtal.box.Ω] = Å³ # Å = 1e-10 m
    # (10^(23) atoms gas / mol gas) * (kg mof / m³) * Å³ * (1e-30 m³ / Å³) = 
    #     1e-7 atoms gas * kg mof / mol gas
    adsorption_scale_factor = 6.02214 * crystal_density(xtal) * xtal.box.Ω * 10^(-7)
    return adsorption_scale_factor
end
```

### Henry Coefficient

```julia
"""
    H = henry_coeff(xtal_name, adsorbate, n_pts; fgroup="")

This function passes the specified number of data points into [fit_adsorption_isotherm](@ref) 
with the `:henry` model option set. 

# Arguments
- `xtal_name::String`: name of the structure the calculation is for
- `adsorbate::String`: the probe molecule
- `n_pts::Int64`: number of points to include in fitting procedure
- `fgroup::String=""`: functional group decorating the organic ligand

# Returns
- `fit::Dict{String, Float64}`: dictionary with estimated Henry Coefficient and associated mean-squared error
"""
function henry_fit(xtal_name::String, adsorbate::String, n_pts::Int64; fgroup::String="")
    # load experimental data
    data = experimental_isotherm_data(adsorbate, xtal_name; fgroup=fgroup)
    fit  = fit_adsorption_isotherm(data[1:n_pts, :], Symbol("pressure (bar)"),
                                     Symbol("⟨N⟩ (mmol/g)"), :henry)
    fit["n_pts"] = Int64(floor(n_pts))
    return fit # the Henry Coefficient
end
```

## Henry Coefficient Fits and Selectivity Plots

Due to the small number of MOF-fgroup-gas combinations considered here, the number of points to include in each fit is tuned manually in each instance.

```julia
# Henry Coefficient
henry_coeff = Dict{Tuple{String, String, String}, Dict{String, Float64}}()

mof="NiPyC2"; fgroup=""; 
henry_coeff[(mof, fgroup, "Xe")] = henry_fit(mof, "Xe", 2)
henry_coeff[(mof, fgroup, "Kr")] = henry_fit(mof, "Kr", 4)
henry_coeff[(mof, fgroup, "Ar")] = henry_fit(mof, "Ar", 4)

mof="NiPyC2"; fgroup="mNH2";
henry_coeff[(mof, fgroup, "Xe")] = henry_fit(mof, "Xe", 2; fgroup=fgroup)
henry_coeff[(mof, fgroup, "Kr")] = henry_fit(mof, "Kr", 4; fgroup=fgroup)
henry_coeff[(mof, fgroup, "Ar")] = henry_fit(mof, "Ar", 4; fgroup=fgroup)

henry_coeff
```

```julia
typeof(henry_coeff[(mof, fgroup, "Ar")]["n_pts"])
```

```julia
# selectivity
selectivity  = Dict{Tuple{String, String, String, String}, Float64}()
for key in keys(henry_coeff)
    for gas in adsorbates
        if key[3] != gas
            selectivity[(key[1], key[2], key[3], gas)] = henry_coeff[key]["H"] / henry_coeff[(key[1], key[2], gas)]["H"]
        end
    end
end
selectivity
```

### Plots for Henry Fits

```julia
# helpful plotting dictionary
henry_linetype  = Dict(("NiPyC2", "") => "--", ("NiPyC2", "mNH2") => "-.")
henry_markers   = Dict(("NiPyC2", "") => "v", ("NiPyC2", "mNH2") => "p")
```

```julia
for fgroup in functional_groups
    figure(figsize=(7, 5))
    for gas in adsorbates
        tup = (parent_mofs[1], fgroup, gas)
        
        # plot Henry Fit (linear)
        p = range(0, stop=1.0, length=100) # define pressure range
        n = henry_coeff[tup]["H"] .* p # apply linear Henry model
        str = @sprintf("%0.2f mmol/(g-bar)", henry_coeff[tup]["H"]) # String for legend
        plot(p, n, label=tup[3] * ", H = " * str,
             color=adsorbate_to_color[tup[3]],
             linestyle=henry_linetype[(tup[1], tup[2])])

        # plot points used in fitting
        n_pts = Int(henry_coeff[tup]["n_pts"])
        scatter(exptl_data[tup][1:n_pts, Symbol("pressure (bar)")], 
                exptl_data[tup][1:n_pts, Symbol("⟨N⟩ (mmol/g)")],
                s=30, 
                label=tup[3] * " (used for fitting)", 
                color=adsorbate_to_color[tup[3]], 
                linewidth=2)

        # plot points not used in fitting
        scatter(exptl_data[tup][n_pts:end, Symbol("pressure (bar)")], 
                exptl_data[tup][n_pts:end, Symbol("⟨N⟩ (mmol/g)")],
                s=30,
                label=tup[3] * " (not used for fitting)", 
                facecolor="none", 
                edgecolor=adsorbate_to_color[tup[3]],
                linewidth=2)

        legend(title="T = 298 K", title_fontsize=10, 
               fontsize=10,  loc=7)

        title(pretty_names[(tup[1], tup[2])], fontsize=16)

    #     xlim(plot_data[tup][:xlims])
    #     ylim(plot_data[tup][:ylims])
        xlim([-0.01, 1.01])
        ylim([-0.1, 3.2])

        xlabel("pressure [bar]")
        ylabel("uptake (mmol/g)")
        tight_layout()

        # save figure
        if save_figures
            filename = "henry_isotherm_fit_$(mof[1])"
            isempty(mof[2]) ? filename *= "" : filename *= "-$(mof[2])"
            savefig(joinpath(pwd(), "henry_coeff", filename * ".pdf"), dpi=600, format="pdf")
        end
    end
end
```

```julia
plt_lims = Dict(
    # parent mof
    ("NiPyC2", "", "Xe")     => Dict(:xlims => [-1E-4, 0.006], :ylims => [-0.01, 0.6]),
    ("NiPyC2", "", "Kr")     => Dict(:xlims => [-1E-4, 0.011], :ylims => [-1E-3, 0.06]),
    ("NiPyC2", "", "Ar")     => Dict(:xlims => [-1E-4, 0.012], :ylims => [-1E-4, 0.006]),
    # functionalized mof
    ("NiPyC2", "mNH2", "Xe") => Dict(:xlims => [-1E-4, 0.006], :ylims => [-0.01, 0.6]),
    ("NiPyC2", "mNH2", "Kr") => Dict(:xlims => [-1E-4, 0.011], :ylims => [-1E-3, 0.06]),
    ("NiPyC2", "mNH2", "Ar") => Dict(:xlims => [-1E-4, 0.012], :ylims => [-1E-4, 0.006])
);
```

```julia
fig, axs = plt.subplots(1, 3, figsize=(15, 5))
for (ind, gas) in enumerate(adsorbates)
    for mof in [("NiPyC2", ""), ("NiPyC2", "mNH2")]
        tup = (mof[1], mof[2], gas)
        
        # plot Henry Fit (linear)
        p = range(0, stop=1.0, length=100) # define pressure range
        n = henry_coeff[tup]["H"] .* p # apply linear Henry model
        str = @sprintf("%0.2f mmol/(g-bar)", henry_coeff[tup]["H"]) # String for legend
        axs[ind].plot(p, n,
                      label=pretty_names[mof] * "\nH = " * str,
                      color=adsorbate_to_color[tup[3]],
                      linestyle=henry_linetype[(tup[1], tup[2])])
        
        # plot points used in fitting
        n_pts = Int(henry_coeff[tup]["n_pts"])
        axs[ind].scatter(exptl_data[tup][1:n_pts, Symbol("pressure (bar)")], 
                         exptl_data[tup][1:n_pts, Symbol("⟨N⟩ (mmol/g)")],
                         s=30, 
                         label="(used for fitting)",
                         marker=henry_markers[(tup[1], tup[2])],
                         color=adsorbate_to_color[tup[3]], 
                         edgecolor="k", linewidth=0.3, Zorder=10)



        axs[ind].set_xlim(plt_lims[tup][:xlims])
        axs[ind].set_ylim(plt_lims[tup][:ylims])
        axs[ind].xaxis.set_major_formatter("{x:,g}") 
    end
        handles, labels = axs[ind].get_legend_handles_labels()
        order = [1, 3, 2, 4]    
        axs[ind].legend([handles[idx] for idx in order], 
                    [labels[idx] for idx in order],
                    title="T = 298 K", title_fontsize=11,
                    fontsize=11, loc="best", edgecolor="none")

        axs[ind].set_xlabel("pressure [bar]")
        axs[1].set_ylabel("uptake (mmol/g)")
        axs[ind].set_title(gas, fontsize=16)
        tight_layout()
end

if save_figures
    filename = "henry_fit_multi-panel"
    savefig(joinpath(pwd(), "henry_coeff", filename * ".pdf"), dpi=600, format="pdf")
end
```

## Adsorption Isotherm Plots

```julia
crystals = Dict{String, Dict{Symbol, Any}}()

crystals["NiPyC2_experiment.cif"] = Dict(:cif   => "NiPyC2_experiment.cif",
                                         :label => "Ni(PyC)"*L"_2",
                                         :syn   => ("NiPyC2", "") # synonym 
                                        )

crystals["Pn_Ni-PyC-NH2.cif"] = Dict(:cif   => "Pn_Ni-PyC-NH2.cif", 
                                     :label => "Ni(PyC-" * L"m" * "-NH" * L"_2" * ")" * L"_2",
                                     :syn   => ("NiPyC2", "mNH2")
                                    )

# crystals["NiPyC2_relax_sc211_meta_functionalized_NH2_pbesol_relax.cif"] = 
#             Dict(:cif   => "NiPyC2_relax_sc211_meta_functionalized_NH2_pbesol_relax.cif",
#                  :label => "Ni(PyC-" * L"m" * "-NH" * L"_2" * ")" * L"_2",
#                  :syn   => ("NiPyC2", "mNH2"))
```

```julia
function plot_single_mof_isotherms(xtal::String, adsorbates::Array{String, 1})
    fig = figure()
    for gas in adsorbates
        df_sim = read_sim_data(props_we_want, sim_params, gas; xtal=crystals[xtal][:cif])

        plot(df_sim[:, Symbol("pressure (bar)")],
             df_sim[:, Symbol("⟨N⟩ (mmol/g)")],
             marker=data_marker[(gas, "sim")],
             label=data_label[(gas, "sim")],
             markersize=8,
             markeredgecolor="black",
             markeredgewidth=0.1,
             color=adsorbate_to_color[gas],
             clip_on=false)

        parent = crystals[xtal][:syn][1]
        fgroup = crystals[xtal][:syn][2]
        scatter(exptl_data[(parent, fgroup, gas)][:,Symbol("pressure (bar)")],
                exptl_data[(parent, fgroup, gas)][:, Symbol("⟨N⟩ (mmol/g)")],
                marker=data_marker[(gas, "exp")],
                label=data_label[(gas, "exp")],
                clip_on=false)

        legend(loc="upper right", fontsize="xx-small", edgecolor="none",
               title= crystals[xtal][:label] * "  @  T = 298 K",
               title_fontsize="x-small", ncol=2)

        grid("on")
        xlabel("pressure [bar]")
        ylabel("uptake [mmol/g]")

        ymin = -0.04
        ymax = 4.5
        ylim(ymin, ymax)

        xmin = -0.01
        xmax = sim_params["pressures"][end] + 0.02
        xlim(xmin, xmax)
    end
    
    # plot commenerate adsorption line
    ads_scale_factor = adsorption_scale_factor(xtal)
    @info ads_scale_factor
    axhline(y=2/ads_scale_factor, linestyle="--", color="0.4")
    text(0.02, 2/ads_scale_factor+0.04, "2 adsorbates/unit cell", color="0.4", fontsize=9)
    
    tight_layout()
    return fig
end
```

```julia
for xtal in ["NiPyC2_experiment.cif", "Pn_Ni-PyC-NH2.cif"]
    plot_single_mof_isotherms(xtal, adsorbates)
end
```
